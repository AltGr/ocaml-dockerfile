[Docker](http://docker.com) is a Linux container manager that can build images
automatically by reading the instructions from a `Dockerfile`. A Dockerfile is
a text document that contains all the commands you would normally execute
manually in order to build a Docker image. By calling `docker build` from your
terminal, you can have Docker build your image step-by-step, executing the
instructions successively.  Read more at <http://docker.com>

This library provides a typed OCaml interface to generating Dockerfiles
programmatically without having to resort to lots of shell scripting and
awk/sed-style assembly.

You can experiment with the library interactively on this page by following
the code cells below.
**`<SHIFT-ENTER>` will evaluate the contents of a cell and advance you to 
the next statement.  Each cell remembers the results of all the previous
evaluations.**

```ocaml
open Dockerfile
```

For the purposes of this tutorial, we will open up the entire `Dockerfile` module.  You can [browse](http://perscon.net/ocaml-dockerfile/api/Dockerfile.html) this module's documentation to see the complete set of functions.

```ocaml
let base =
  let email = "anil@recoil.org" in
  comment "Generated by OCaml Dockerfile" @@
  from "ubuntu" ~tag:"trusty" @@
  maintainer "Anil Madhavapeddy <%s>" email
  
```

Every Dockerfile must begin with a `FROM` and `MAINTAINER` instruction to indicate the base image and who to send feedback to.  The `base` variable now records these instruction as a list, with the `@@` operator concatenating individual instructions.

```ocaml
string_of_t base
```

`Dockerfile.string_of_t` converts an instruction stream into a string in the Dockerfile format. You can print this out to a file using the usual OCaml commands (for example, `open_out` followed by `output_string`).

```ocaml
let run_as_user user fmt = 
  Printf.ksprintf (run "sudo -u %s sh -c %S" user) fmt ;;
```

```ocaml
string_of_t (run_as_user "opam" "opam update -u -y")
```

You can also wrap the base functions in higher-level format strings that abstract away concerns such as running under `sudo` as a different user.

```ocaml
let ocaml_ubuntu_image =
  base @@
  run "apt-get -y -qq update" @@
  run "apt-get -y install ocaml ocaml-native-compilers camlp4-extra opam" @@
  onbuild (run "apt-get -y -qq update") ;;
```

```ocaml
print_endline (string_of_t ocaml_image)
```

The above two code blocks tie together the previous definitions, along with a recursive `ONBUILD` instruction which triggers an update when another image is constructed from this one.  Finally, we print out a complete `Dockerfile` to the console.
